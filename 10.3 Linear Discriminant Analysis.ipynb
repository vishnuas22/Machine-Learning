```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Load Iris Dataset

data = load_iris()

X = data.data

y = data.target

class_labels = data.target_names
```

```python
# Apply LDA

lda = LinearDiscriminantAnalysis(n_components=2)

X_lda = lda.fit_transform(X,y)
```

```python
# Visualize the LDA-transformed data

plt.figure(figsize=(8,6))

for class_value,class_label in enumerate(class_labels):
    plt.scatter(X_lda[y==class_value,0], X_lda[y==class_value,1], label=class_label)

plt.title("LDA on Iris Dataset")
plt.xlabel("LDA component 1")
plt.xlabel("LDA component 2")
plt.legend()
plt.grid()
plt.show()
```
