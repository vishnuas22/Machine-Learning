```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing # Custom Dataset from sklearn for practice
```

```python
# Load the California Housing dataset

california = fetch_california_housing(as_frame = True)
df = california.frame
```

```python
# Add the target (median house value) to the DataFrame

df['MedHouseVal'] = df['MedHouseVal']
```

```python
# Calculate correlation

correlation_matrix = df.corr()
```

```python
# Visualize the heatmap

plt.figure(figsize=(10,5))

sns.heatmap(correlation_matrix, annot= True, cmap='coolwarm', fmt='.2f')

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Select features with high correlation with target

target_corr = correlation_matrix['MedHouseVal'].abs().sort_values(ascending = False)

print( 'Correlation with Target Variables (MedHouseVal) \n', target_corr)
```

```python
# Filter features with |correlation| > 0.5

selected_features = target_corr[target_corr > 0.5].index.tolist()

print('\n Selected features based on correlation \n',selected_features)
```
