```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
```

```python
# Generate synthetic data
data = {
    "Feature": [0.01, 0.05, 0.2, 0.4, 0.8, 1.2, 2.5, 5.0, 10.0],
    "Target": [1, 2, 3, 4, 5, 6, 7, 8, 9]  # Simulated linear relationship
}
```

```python
df = pd.DataFrame(data)
```

```python
# Log Transformation
df["Log_feature"] = np.log(df["Feature"] + 1e-5) #Adding a small value to avoid log(0)
```

```python
# Sqaure root Transformation
df["sqrt_feature"] = np.sqrt(df["Feature"] ) 
```

```python
# Min-Max scaling
scaler = MinMaxScaler()
df["Min_max_feature"] = scaler.fit_transform(df[["Feature"]])
```

```python
# Visualization

plt.figure(figsize=(12,6))

plt.subplot(1,3,1)

plt.plot(df["Feature"], df["Log_feature"], marker='o', label = "Log Transform")

plt.title("Log Transformation")

plt.xlabel("Feature")

plt.ylabel("Transformed Feature")

plt.legend()

plt.tight_layout()
plt.show()
```

```python

plt.subplot(1,3,2)

plt.plot(df["Feature"], df["sqrt_feature"], marker='o', color="orange" ,label = "Sqrt Transform")

plt.title("Sqrt Transformation")

plt.xlabel("Feature")

plt.ylabel("Transformed Feature")

plt.legend()

plt.tight_layout()
plt.show()
```

```python

plt.subplot(1,3,3)

plt.plot(df["Feature"], df["Min_max_feature"], marker='o', color="green", label = "Min_max Scaling")

plt.title("Min max scaling")

plt.xlabel("Feature")

plt.ylabel("Transformed Feature")

plt.legend()

plt.tight_layout()
plt.show()
```
