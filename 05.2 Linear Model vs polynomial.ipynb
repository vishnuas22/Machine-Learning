```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error
```

```python
# Original housing dataset
data = {
    "Size (sqft)": [2000, 1500, 1800, 1200, 2500],
    "Bedrooms": [3, 2, 3, 2, 4],
    "Bathrooms": [2, 1, 2, 1, 3],
    "Location Score": [8, 7, 9, 6, 10],
    "Price ($)": [300000, 200000, 250000, 180000, 500000]
}

```

```python
# Convert to Dataframe

df = pd.DataFrame(data)
```

```python
# Extract features and target
X = df[["Size (sqft)"]]
y = df["Price ($)"]
```

```python
# Fit a linear regression model
linear_model = LinearRegression()
linear_model.fit(X, y)
y_pred_linear = linear_model.predict(X)
```

```python
# Add Polynomial features( degree=2 for quadratic)
poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)
```

```python
# Fit a Polynomial Regression model

poly_model = LinearRegression()
poly_model.fit(X_poly, y)
y_pred_poly = poly_model.predict(X_poly)
```

```python
# Visualize the results

plt.figure(figsize=(10, 6))
plt.scatter(data["Size (sqft)"], data["Price ($)"], color='blue', label="Data points")
plt.plot(data["Size (sqft)"], y_pred_linear, color='red', label="Linear Regression")
plt.plot(data["Size (sqft)"], y_pred_poly, color='green', label="Polynomial Regression (Degree 2)")
plt.xlabel("Size (sqft)")
plt.ylabel("Price ($)")
plt.title("Linear vs Polynomial Regression")
plt.legend()
plt.show()
```

```python
# Compare Mean Squared Errors
mse_linear = mean_squared_error(y, y_pred_linear)
mse_poly = mean_squared_error(y, y_pred_poly)
print(f"Linear Regression MSE: {mse_linear:.2f}")
print(f"Polynomial Regression MSE: {mse_poly:.2f}")
```
