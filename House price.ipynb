```python
import numpy as np

from sklearn.linear_model import LinearRegression

import matplotlib.pyplot as plt
```

```python
sizes = np.array([[1000],[1500],[2000],[2500],[3000]])

prices = np.array([200000,250000, 300000,350000,400000])
```

```python
model = LinearRegression()

model.fit(sizes,prices)
```

```python
predicted_price = model.predict([[2200]])

print( f" Predicted price for a 2200 sqft House : {predicted_price[0]:,.2f}")
```

```python
plt.scatter(sizes,prices,color='purple')
plt.plot(sizes,model.predict(sizes),color='red')
plt.xlabel('Sizes (sqft)',color='violet')
plt.ylabel('Prices (USD)',color='violet')
plt.title('House price prediction',color='violet')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 2D column vector for sizes
sizes_2d = np.array([[1000],[1500],[2000],[2500],[3000]])

# 1D row vector for sizes (alternative representation)
sizes_1d = np.array([1000, 1500, 2000, 2500, 3000])

# Prices as 1D array
prices = np.array([200000, 250000, 300000, 350000, 400000])

# Linear Regression with 2D sizes
reg_2d = LinearRegression().fit(sizes_2d, prices)
print("2D Sizes Regression:")
print("Coefficient:", reg_2d.coef_)
print("Intercept:", reg_2d.intercept_)

# Linear Regression with 1D sizes
reg_1d = LinearRegression().fit(sizes_1d.reshape(-1, 1), prices)
print("\n1D Sizes Regression:")
print("Coefficient:", reg_1d.coef_)
print("Intercept:", reg_1d.intercept_)
```
