```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a Skewed Dataset

data = {

    "Age" : [22, 25, 30, 35, 40, 50, 60],
    "Income": [15000, 18000, 30000, 60000, 120000, 250000, 500000]
}
```

```python
# Convert to dataframe

df = pd.DataFrame(data)
```

```python
# Visualize the original distribution

plt.figure(figsize=(10,4))

plt.subplot(1, 2, 1)

plt.hist(df['Income'], bins=10, color='skyblue', edgecolor ='black')

plt.title('Original income distribution')

plt.xlabel('Income')

plt.ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
# Apply log Transformation

df['log_income'] = np.log(df['Income'])
```

```python
# Visualize the transformed distribution

plt.subplot(1, 2, 2)

plt.hist(df['log_income'],bins=10, color='orange', edgecolor='black')

plt.title('Log-Transformed Income Distribution')

plt.xlabel('Log(Income)')

plt.ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
# Display the transformed Dataset

print(df)
```
