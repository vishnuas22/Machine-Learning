Use Case: Detecting network intrusions & cyber attacks
Why DBSCAN? It can detect outliers (attack patterns) that donâ€™t belong to normal clusters.

```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN

from sklearn.preprocessing import StandardScaler
```

```python
# Simulated network traffic (data format: [bytes_sent, bytes_received])

X = np.array([

    [50,10],[60,12],[55,9],[70,15],[90,30], # Normal traffic
    [300,500],[400,600],[1000,2000],[1500,3000]  # Anomalies (DDoS or malware)

])
```

```python
# Standardize the data

X_scaled = StandardScaler().fit_transform(X)
```

```python
# Apply DBSCAN

dbscan = DBSCAN(eps=1, min_samples=2).fit(X_scaled)

labels = dbscan.labels_
```

```python
# Plot results

plt.figure(figsize=(10,7))

plt.scatter(X[:,0],X[:,1],c=labels, cmap='viridis',edgecolors='k', s=100)

plt.xlabel('Bytes Sent')

plt.ylabel('Bytes Recieved')

plt.title("Network Traffic Anomaly Detection using DBSCAN")


# Highlight anomalies (label = -1)

for i, txt in enumerate(labels):

    if txt== -1:
        plt.annotate('Anomaly',(X[i,0], X[i,1]),textcoords='offset points',xytext=(5,5), ha='center')


plt.show()
```

 Outcome: Anomalies (potential cyber attacks) are identified as outliers (label -1).
