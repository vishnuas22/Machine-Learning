DBSCAN (Density-Based Spatial Clustering of Applications with Noise)


DBSCAN is different from K-Means and Hierarchical Clustering because:
It doesn’t require specifying the number of clusters beforehand.
It identifies noise points (outliers).
It works well for arbitrarily shaped clusters.

 How DBSCAN Works
 
1️⃣ Defines core points: If a point has at least min_samples neighbors within eps (radius), it's a core point.
2️⃣ Expands clusters: Core points spread to density-reachable points.
3️⃣ Marks noise points: If a point is not reachable from any core point, it's classified as noise.

```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import make_moons

from sklearn.cluster import DBSCAN

from sklearn.preprocessing import StandardScaler
```

```python
# Generate a synthetic dataset with two moon-shaped clusters

X, _ = make_moons(n_samples=500, noise=0.1, random_state=42)
```

```python
# Standardize features (DBSCAN is sensitive to different scales)

X = StandardScaler().fit_transform(X)
```

```python
# Apply DBSCAN

dbscan = DBSCAN(eps=0.3, min_samples=5)    # eps = neighborhood radius, min_samples = minimum points for a cluster

labels = dbscan.fit_predict(X)
```

```python
# Plot the results

plt.figure(figsize=(8,5))

unique_labels = set(labels)

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
```

```python
for label,color in zip(unique_labels,colors):
    
    mask = (labels == label)

    plt.scatter(X[mask,0],X[mask,1], label = f'Cluster {label}' if label != -1 else 'Noise', alpha=0.7, edgecolors='k')


plt.title("DBSCAN Clustering")

plt.legend()

plt.show()
```

✅ DBSCAN automatically detects clusters without needing K.
✅ Noise points are marked with -1.
✅ Works well for non-spherical clusters like moons & spirals.
