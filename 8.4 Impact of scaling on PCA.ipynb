```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler

from sklearn.decomposition import PCA
```

```python
# Step 1: Create synthetic data with different scales

np.random.seed(42)

feature_1 = np.random.rand(100) * 10  # Feature 1: Range 0-10
feature_2 = np.random.rand(100) * 1000  # Feature 2: Range 0-1000
feature_3 = np.random.rand(100) * 10000  # Feature 2: Range 0-1000
```

```python
data = pd.DataFrame({'Feature1': feature_1, 'Feature2': feature_2, 'Feature3': feature_3})
```

```python
# Step 2: Apply PCA without scaling

pca = PCA()
pca_without_scaling = pca.fit_transform(data)
explained_variance_no_scaling = pca.explained_variance_ratio_
```

```python
# Step 3: Scale the data and reapply PCA

scaler = StandardScaler()
scaler.fit_transform(data)
pca_scaled = PCA()
pca_with_scaling = pca_scaled.fit_transform(data)
explained_variance_scaling = pca_scaled.explained_variance_ratio_
```

```python
# Step 4: Visualize explained variance ratios

x = np.arange(1,len(explained_variance_no_scaling) + 1)

plt.figure(figsize=(10,6))
plt.bar( x - 0.2, explained_variance_no_scaling, width = 0.4, label = 'Without Scaling')
plt.bar( x + 0.2, explained_variance_scaling, width = 0.4, label = 'With Scaling')

plt.xlabel('Principal components')
plt.ylabel('Explained Variance ratio')
plt.title('Effect of scaling in PCA')
plt.legend()
plt.show()
```
