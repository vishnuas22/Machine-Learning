```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data

np.random.seed(42)
size = np.random.uniform(500, 3000, 50) # House size in the sqft
noise = np.random.normal(0, 10000, 50) # Random noise
price = 0.5 * size**2 - 100 * size + 5000 + noise # Quadratic realtionship
```

```python
# Create a Dataframe

data = pd.DataFrame({"Size (sqft)": size,"Price ($)": price })
```

```python
# Scatter plot of the data

plt.figure(figsize=(10,6))
plt.scatter(data["Size (sqft)"],data["Price ($)"], color='green', label = "Data points")
plt.xlabel("Size (sqft)")
plt.ylabel("Price ($)" )
plt.title("House prices v/s Size (Non-Linear Relationship)")
plt.tight_layout()
plt.legend()
plt.show()
```
