Use Case: Grouping customers based on purchasing behavior.

Why DBSCAN? It finds dense clusters of similar spending patterns without needing a predefined number of groups.

```python
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

from sklearn.preprocessing import StandardScaler

from sklearn.cluster import DBSCAN
```

```python
# Simulate customer spending data

X, _ = make_blobs(n_samples=300, centers=[[200, 5000],[600, 10000],[1000, 15000]], cluster_std=100, random_state=42)

# X, _ = make_blobs(n_samples=300, centers=[[200, 5000], [600, 10000], [1000, 15000]], cluster_std=100, random_state=42)
```

```python
# Standardize and apply DBSCAN

X_scaled = StandardScaler().fit_transform(X)
```

```python
# Initialize Model and seperate labels

dbscan = DBSCAN(eps=0.5, min_samples=5).fit(X_scaled)

labels = dbscan.labels_
```

```python
# Plot clusters

plt.scatter(X[:,0],X[:,1],c=labels,cmap='viridis', edgecolors='k', s=50)

plt.xlabel('Purchase Frequency')

plt.ylabel('Total Spending $')

plt.title('Customer Segmentation Using DBSCAN')

plt.show()
```
