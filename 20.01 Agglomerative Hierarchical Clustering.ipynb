```python
import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

import matplotlib.patches as mpatches

import seaborn as sns

from sklearn.datasets import make_blobs

from sklearn.cluster import AgglomerativeClustering

from scipy.cluster.hierarchy import dendrogram, linkage

```

```python
# Generate synthetic dataset

X,_ = make_blobs(n_samples=200, centers=4, cluster_std=1.2 , random_state=42)
```

```python
# Step 1: Plot the dendrogram to determine the optimal number of clusters

plt.figure(figsize=(19,12))

Z = linkage(X, method='ward')  # Ward's method minimizes variance within clusters

dendrogram(Z)

plt.title('Dendrogram For Hierarchical Clustering')

plt.xlabel('Data Points')

plt.ylabel('Distance')

plt.show()
```

Step 2: Apply Agglomerative Clustering


```python
num_clusters = 4  # Based on dendrogram analysis

agg_clustering = AgglomerativeClustering(n_clusters=num_clusters, linkage='ward')

labels = agg_clustering.fit_predict(X)
```

```python
# Step 3: Visualize the Clusters

plt.figure(figsize=(8,5))

unique_gmm_labels = np.unique(labels)

gmm_cmap = plt.get_cmap('tab10')

plt.scatter(x=X[:,0],y=X[:,1], c=labels, edgecolors='k', cmap='tab10', alpha=0.7)

# Create legend handles for GMM Clustering

gmm_legend_handles = [mpatches.Patch(color=gmm_cmap(i / max(unique_gmm_labels)), label=f'Cluster {i}') for i in unique_gmm_labels]

plt.legend(handles=gmm_legend_handles, title="Clusters")

plt.xlabel('Feature 1')

plt.ylabel('Feature 2')

plt.show()
```
