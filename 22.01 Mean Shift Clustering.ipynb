```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs

from sklearn.cluster import MeanShift,estimate_bandwidth
```

```python
# Generate sample data

X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=42)
```

```python
# Estimate bandwidth (automatically chooses a good radius for clusters)

bandwidth = estimate_bandwidth(X, quantile=0.2)
```

```python
# Apply Mean Shift

meanshift = MeanShift(bandwidth=bandwidth)

labels = meanshift.fit_predict(X)
```

```python
# Plot results

plt.scatter(X[:,0], X[:,1], c=labels, cmap='tab10',edgecolors='k', marker='o')

plt.scatter(meanshift.cluster_centers_[:,0], meanshift.cluster_centers_[:,1], c='White', s=200, marker='*', label = 'Centers')

plt.title('Mean Shift Clustering')

plt.legend()

plt.show()
```

 No need to specify the number of clusters (K), but it's slow on large datasets.
