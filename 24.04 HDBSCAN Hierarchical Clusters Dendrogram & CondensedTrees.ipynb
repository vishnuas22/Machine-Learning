One of the biggest advantages of HDBSCAN over traditional clustering is its hierarchical structure, which allows us to visualize how clusters form at different levels.

Instead of just flat clusters, we can:

‚úÖ Analyze the cluster hierarchy to find the best cluster structure.

‚úÖ Decide the optimal granularity (fine vs. coarse clustering).

‚úÖ Manually adjust clustering based on domain knowledge.

üîπ How Does HDBSCAN‚Äôs Hierarchical Clustering Work?

1Ô∏è‚É£ It first builds a hierarchical tree of clusters based on density.

2Ô∏è‚É£ Then, it condenses the tree by removing weak links, leaving only strong clusters.

3Ô∏è‚É£ Finally, it extracts flat clusters at an optimal level in the hierarchy.

üõ† Tools for Hierarchical Visualization in HDBSCAN:

‚úÖ Dendrograms (Tree structure showing how clusters merge/split).

‚úÖ Condensed Trees (More compact & meaningful visualization).

```python
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

import hdbscan

from sklearn.datasets import make_blobs
```

```python
# Generate synthetic dataset with multiple density clusters

X, _ = make_blobs(n_samples=500, centers=[[0,0],[5,5],[10,0]], cluster_std=[0.5, 1.2, 0.8], random_state=42)
```

```python
# Apply HDBSCAN clustering

clusterer = hdbscan.HDBSCAN(min_cluster_size=10, 
                      min_samples=5, 
                      gen_min_span_tree=True
                      
                      )

cluster_labels = clusterer.fit_predict(X)

```

```python
# Plot condensed tree

plt.figure(figsize=(10,6))

clusterer.condensed_tree_.plot(select_clusters=True, selection_palette=sns.color_palette())

plt.title("HDBSCAN Condensed Tree (Hierarchical Cluster Structure)")

plt.show()
```

üîπ Understanding the Condensed Tree Visualization

üîπ The tree starts from a single cluster at the top and splits down.

üîπ Each branch represents a potential cluster, and we can choose different levels.

üîπ The algorithm automatically selects the best cut (marked in color).

Dendrogram Visualization

We can also manually explore different clustering levels using a dendrogram, like this:

```python
# Plot Dendrogram

plt.figure(figsize=(19,11))

clusterer.single_linkage_tree_.plot(cmap='viridis',colorbar=True)

plt.title("HDBSCAN Dendrogram (Cluster Merging Process)")

plt.show()
```

‚úÖ HDBSCAN automatically finds the best cluster structure, but we can manually adjust it.

‚úÖ The condensed tree lets us explore different levels of granularity.

‚úÖ Dendrograms help us understand how clusters merge at different densities.
